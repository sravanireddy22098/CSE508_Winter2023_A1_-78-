```python
import os
import pandas as pd
import nltk
from nltk import word_tokenize
from bs4 import BeautifulSoup
from collections import defaultdict
nltk.download('punkt')

# assign directory
directory = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'

# iterate over files in that directory
doc = 0
def init_postings():
    return {'doc_ids': set(), 'positions': {}}
doc_name_id={}
postings = defaultdict(init_postings)

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    doc_name_id[str(doc)]=filename
    # checking if it is a file
    if os.path.isfile(f):
        read_data = open(f, "r")
        contents = read_data.read().strip()
        if doc % 50 == 0:
            print(doc)
        # create a list of unigrams from the text
        words = word_tokenize(contents)
        for i, word in enumerate(words):
            if word in postings:
                if doc not in postings[word]['doc_ids']:
                    postings[word]['doc_ids'].add(doc)
                postings[word]['positions'].setdefault(doc, []).append(i)
            else:
                postings[word]['doc_ids'].add(doc)
                postings[word]['positions'][doc] = [i]
        doc += 1

# print the unigram positional index
for word, data in postings.items():
    print(f'{word}: {{"doc_freq": {len(data["doc_ids"])}, "positions": {data["positions"]}}}')
    
# save the unigram positional index to a pickle file
#print(postings)

```

```python
def XANDY(x, y):
    x, y = sorted(x), sorted(y)
    _AND = []
    comp = 0
    i, j = 0, 0
    while i<len(x) and j<len(y):
        if x[i] == y[j]:
            _AND.append(y[j])
            j += 1
            i += 1
            comp += 1
        else:
            comp += 2
            if x[i]<y[j]:
                i+=1
            else:
                j+=1
    return _AND, comp #returning the and array and the number of comparisons we made
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
def preprocess(input):
  txt_l = input.lower()
  txt_tk = nltk.word_tokenize(txt_l)
  str_tk = ' '.join(str(token) for token in txt_tk)
  stopwords_ = stopwords.words('english')
  tokens = nltk.word_tokenize(str_tk)
  tokens_without_stopwords = [token for token in tokens if token.lower() not in stopwords_]
  text_without_stopwords = ' '.join(tokens_without_stopwords)
  txt_punc = re.sub(r'[^\w\s]', '', text_without_stopwords)
  txt_b = txt_punc.replace("  ", " ")
  return txt_b
```

```python
from nltk import word_tokenize
N = int(input("Enter number of queries:"))
queries = []
for i in range(N):
  print("Query ",i+1,":",end=" ")
  x = input()
  queries.append(x)

cnt = 0

for query in queries:
  wl=preprocess(query)
  #print(wl)
  wl = word_tokenize(wl)
  list_of_tup = []
  for i in range(len(wl)-1):
      list_of_tup.append((wl[i],wl[i+1]))
  print(list_of_tup)
  process(list_of_tup,cnt+1)
  cnt += 1
```
