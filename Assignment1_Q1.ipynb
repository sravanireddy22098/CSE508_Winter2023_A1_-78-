1.Data Preprocessing

i).Text Extraction.

```python
import os
from bs4 import BeautifulSoup

dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'
i = 1
for filename in os.listdir(dire):
    f = os.path.join(dire, filename)
    file1 = open(f)
    soup = file1.read()
    bs = BeautifulSoup(soup, "html.parser")
    title_tag = bs.find('title')
    text_tag = bs.find('text')
    if title_tag is None or text_tag is None:
        print(f"Error: {filename} does not contain <title> and <text> tags")
        continue
    title = title_tag.text
    text = text_tag.text
    texti = title + text
    file1.close()
    if i < 6:
        print("*****************"+"file before preprocessing:" + filename+"****************")
        print(soup)
        print("*****************"+"files after preprocessing:" + filename+"*****************")
        print(texti)
    file2 = open(f, 'r+')
    file2.truncate(0)
    file2.close()
    file3 = open(f, 'w')
    file3.write(texti)
    i += 1
    file3.close()
```

ii).preprocessing

1. lower case the text

```python
dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'
i=1
for filename in os.listdir(dire):
    f = os.path.join(dire, filename)
    with open(f, 'r') as file1:
        txt = file1.read()
        txt_l = txt.lower()
        if i<6:
            print("*************""file before lowering case:" + filename+"****************")
            print(txt)
            print("**************""file after lowering case:" + filename+"***************")
            print(txt_l)
    with open(f, 'w') as file2:
        file2.write(txt_l)
    i = i+1
```

2.Perform Tokenization

```python
import os
import nltk
nltk.download('punkt')

dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'
i = 1

for filename in os.listdir(dire):
    f = os.path.join(dire, filename)
    with open(f, 'r') as file1:
        txt = file1.read()
        
    txt_tk = nltk.word_tokenize(txt)
    str_tk = ' '.join(str(token) for token in txt_tk)
    
    if i < 6:
        print("**************"+"files before tokenization:" + filename+"******************")
        print(txt)
        print("\n")
        print("*****************"+"file after tokenization:" + filename+"*******************")
        print(txt_tk)
        print(type(txt_tk))
        print("\n")
    
    with open(f, 'w') as file2:
        file2.write(str_tk)
        
    i += 1
```

3.Remove Stop Words

```python
import nltk
from nltk.corpus import stopwords
import os

stopwords = stopwords.words('english')
dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'

for i, filename in enumerate(os.listdir(dire)):
    if i >= 5:
        break

    file_path = os.path.join(dire, filename)

    with open(file_path, 'r') as file:
        text = file.read()
        tokens = nltk.word_tokenize(text)
        tokens_without_stopwords = [token for token in tokens if token.lower() not in stopwords]
        text_without_stopwords = ' '.join(tokens_without_stopwords)

        print("\n****************** with stop words: " + filename + " ********************")
        print(text)
        print("\n****************** After stop words removal: " + filename + " ******************")
        print(text_without_stopwords)

    with open(file_path, 'w') as file:
        file.write(text_without_stopwords)
```

4.Remove punctuations

```python
import re
import os

dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'
i = 1

for filename in os.listdir(dire):
    f = os.path.join(dire, filename)
    with open(f, 'r') as file1:
        txt = file1.read()
        txt_punc = re.sub(r'[^\w\s]', '', txt)

    if i <= 5:
        print("\n")
        print("*****************" + "file before removing punctuations:" + filename + "*****************")
        print(txt)
        print("\n")
        print("*****************" + "file after removing punctuations:" + filename + "******************\n")
        print(txt_punc)

    with open(f, 'w') as file3:
        file3.write(txt_punc)

    i += 1
```

5.Remove blank spaces

```python
dire = '/content/drive/MyDrive/CSE508_Winter2023_Dataset'
i = 1

for filename in os.listdir(dire):
    f = os.path.join(dire, filename)
    with open(f, 'r') as file1:
        txt = file1.read()
        txt_b = txt.replace("  ", " ")
    if i <= 5:
        print("\n")
        print("*****************" + "file before removing blank spaces: " + filename + "*****************")
        print(txt)
        print("\n")
        print("*****************" + "file after removing blank spaces: " + filename + "******************\n")
        print(txt_b)
    with open(f, 'r+') as file2:
        file2.truncate(0)
    with open(f, 'w') as file3:
        file3.write(txt_b)
    i += 1
```
